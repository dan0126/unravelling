```python
df = pd.read_csv('../input/dados-das-interfaces-1/H.dat',sep="\s+", decimal='.')


df.columns = ['Interfaces', 'Propriedades','Fluidos','Valores']

prop = df["Propriedades"].unique().tolist()
interf = df["Interfaces"].unique().tolist()

print("Quantidade de tipos de propriedades sem repetição: ",len(prop))
print("Quantidade de tipos de interfaces sem repetição: ",len(interf))
```

**Óleo**

```python
df_oil = df[df['Fluidos'] == "OIL"]

table_oil = pd.pivot_table(data = df_oil, index = ["Interfaces"], columns = ["Propriedades"])
#pd.set_option("display.max_rows", None, "display.max_columns", None)

table_oil.reset_index(level=0, inplace=True)

table_oil.shape
```

```python
table_oil
```

```python
prop_oil = df_oil["Propriedades"].unique().tolist()

def dot_to_comma(df,j):
    
    l = df.iloc[:,j].tolist()
    
    new = []
    
    for i in range(len(l)):
            
        new.append(str(l[i]).replace('.', ','))
            
    df.iloc[:,j] = new

for j in range(len(prop_oil)):

    dot_to_comma(table_oil,j)  
    
table_oil #substituído ponto por vírgula
```

```python
table_oil.to_csv('table_oil.csv',index=False)
```

#  Eliminação das colunas com 'nan' e Transformação dos dados em string para float 

**Heptane**

```python
toluene_curt = []

for text in prop_toluene:
    
    if 'CUR' in text:
        
        toluene_curt.append(text)
```

```python
df_curt_tol = toluene.filter(toluene_curt, axis=1)

df_curt_tol
```

```python
def cont_nan(df,colum_name):
    
    cont=0
    
    for j in range(len(df[colum_name])):
        
        if df[colum_name][j] == 'nan':
        
            cont += 1
     
    if cont >= 15:
        
        return colum_name
```

```python
for i in toluene_curt:
    
    name = cont_nan(df_curt_tol,i)
    
    if name != None:
    
        df_curt_tol = df_curt_tol.drop(name, axis=1)
```

```python
for i in df_curt_tol.columns.to_list():
    
    for j in range(len(df_curt_tol[i])):
    
        if df_curt_tol[i][j] == 'nan':
            
            df_curt_tol[i][j] = '0'
```

```python
for i in df_curt_tol.columns.to_list():
    
    df_curt_tol[i] = df_curt_tol[i].astype(float)
```

```python
df_curt_tol.mean()
```

```python
df_curt_tol.mean().plot()

plt.xticks(rotation="30")
```

# Aprendizado de máquina - Feature Importance

https://towardsdatascience.com/3-essential-ways-to-calculate-feature-importance-in-python-2f9149592155

```python
import sklearn
from pylab import rcParams
```

**HEPTANO**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X = heptane.drop('IFT', axis=1)
y = heptane['IFT']

for i in range(len(y)):
    
    y[i] = np.round(y[i])
    
y = y.astype(int)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

ss = StandardScaler()

X_train_scaled = ss.fit_transform(X_train)
X_test_scaled = ss.transform(X_test)
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train_scaled, y_train)
importances = pd.DataFrame(data={
    'Attribute': X_train.columns,
    'Importance': model.coef_[0]
})
importances = importances.sort_values(by='Importance', ascending=False)
```

```python
rcParams['figure.figsize'] = 20, 10

plt.bar(x=importances['Attribute'], height=importances['Importance'], color='#087E8B')
plt.title('[LGR] Heptano - Feature importances obtained from coefficients', size=20)
plt.xticks(rotation='vertical')
plt.show()
```

```python
from xgboost import XGBClassifier

model = XGBClassifier()
model.fit(X_train_scaled, y_train)
importances = pd.DataFrame(data={
    'Attribute': X_train.columns,
    'Importance': model.feature_importances_
})
importances = importances.sort_values(by='Importance', ascending=False)
```

```python
rcParams['figure.figsize'] = 20, 10

plt.bar(x=importances['Attribute'], height=importances['Importance'], color='#087E8B')
plt.title('[XGB Classifier] Heptano - Feature importances obtained from coefficients', size=20)
plt.xticks(rotation='vertical')
plt.show()
```
