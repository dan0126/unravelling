# Códigos para tratamento de dados

```python
def comma_to_dot(df,j):
    
    l = df.iloc[:,j].tolist()
    
    new = []
    
    for i in range(len(l)):
            
        new.append(float(str(l[i]).replace(',', '.')))
            
    df.iloc[:,j] = new
```

```python
light = pd.read_csv('../input/ift-com-propriedades-completo/TABLE2.xlsx - Light.csv')
toluene = pd.read_csv('../input/ift-com-propriedades-completo/TABLE2.xlsx - Toluene.csv')
heptane = pd.read_csv('../input/ift-com-propriedades-completo/TABLE2.xlsx - Heptane.csv')
heptol = pd.read_csv('../input/ift-com-propriedades-completo/TABLE2.xlsx - Heptol.csv')
medium = pd.read_csv('../input/ift-com-propriedades-completo/TABLE2.xlsx - Medium.csv')


for j in range(len(light.columns.tolist())):

    comma_to_dot(light,j)  #substituído ponto por vírgula
    
for j in range(len(toluene.columns.tolist())):

    comma_to_dot(toluene,j) #substituído ponto por vírgula
    
for j in range(len(heptane.columns.tolist())):

    comma_to_dot(heptane,j)  #substituído ponto por vírgula
    
for j in range(len(heptol.columns.tolist())):

    comma_to_dot(heptol,j) #substituído ponto por vírgula
    
for j in range(len(medium.columns.tolist())):

    comma_to_dot(medium,j) #substituído ponto por vírgula
    
light = light.dropna(axis=1, how='all')
toluene = toluene.dropna(axis=1, how='all')
heptane = heptane.dropna(axis=1, how='all')
heptol = heptol.dropna(axis=1, how='all')
medium = medium.dropna(axis=1, how='all')

light = light.fillna(0)
toluene = toluene.fillna(0)
heptane = heptane.fillna(0)
heptol = heptol.fillna(0)
medium = medium.fillna(0)
```

# Machine Learning - feature importance

```python
import sklearn
from pylab import rcParams
import seaborn as sns
import matplotlib.pyplot as plt
```

**HEPTANO**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X = heptane.drop('IFT', axis=1)
y = heptane['IFT']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from xgboost import XGBRegressor

model = XGBRegressor(random_state=0)
model.fit(X_train, y_train)
importances = pd.DataFrame(data={
    'Attribute': X_train.columns,
    'Importance': model.feature_importances_
})
importances = importances.sort_values(by='Importance', ascending=False)

importances = importances[importances['Importance'] >= 0.01]
```

```python
# example of calculate the mean absolute error
from sklearn.metrics import mean_absolute_error

# make predictions for test data
predicted = model.predict(X_test)

# calculate errors
errors = mean_absolute_error(y_test, predicted)
# report error
print(errors)
```

```python
rcParams['figure.figsize'] = 5, 5

plt.bar(x=importances['Attribute'], height=importances['Importance'], color='#087E8B')
plt.title('[XGB Classifier] Heptano - Feature importances obtained from coefficients', size=20)
plt.xticks(rotation='vertical')
plt.show()
```
